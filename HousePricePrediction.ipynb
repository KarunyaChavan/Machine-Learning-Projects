Step 1 - Import Required Libraries

```python
import pandas as pd
```

Step 2 - Import and Read the data.

```python
houses = pd.read_csv("https://raw.githubusercontent.com/YBI-Foundation/Dataset/main/House%20Prices.csv")
```

```python
houses.columns
```

Run three basic functions on the dataset which has been imported

```python
houses.head()
```

```python
houses.info()
```

```python
houses.describe()
```

Step 3 - Define Target(y) and Features(x)

```python
houses.shape
```

```python
y = houses['Price']
x = houses[['Bedrooms', 'Bathrooms', 'Sqft_living',
       'Sqft_lot', 'Floors', 'Waterfront', 'View', 'Condition', 'Grade',
       'Sqft_above', 'Sqft_basement', 'Yr_built', 'Yr_renovated', 'zipcode',
       'Lat', 'Long', 'Sqft_living15', 'Sqft_lot15']]
```

```python
x.shape
```

```python
y.shape
```

Step 4 - Train, Test and Spilt the Data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size = 0.7, random_state=2529)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
x_train
```

Step 5 - Select the Model (Linear Regression)

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

Step 6 - Train the Model

```python
model.fit(x_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

Step 7 - Prediction

```python
y_pred = model.predict(x_test)
```

```python
x_test
```

Step 8 - Accuracy Testing

```python
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_test,y_pred)
```

Error = 25.6%;
Accuracy = 74.4;

Visualization of Actual VS Predicted Results

```python
import matplotlib.pyplot as plt
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Predicted Price")
plt.show()
```
